## Importing Python Libraries

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from mpl_toolkits import mplot3d
```

## Function with relative minimum

```
#Function with relative min
def min(x,y):
  return x ** 2 + y ** 2
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

#Plotting in a three-dimensional 3d contour plot first
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = min(X, Y)

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

#Marking local minimum
ax.scatter(0,0,0, marker='o', color='r')
```

```
#Better view of the local minimum in the 3d contour plot
ax.view_init(60, 35)
fig
```

```
#Visualizing the function in a 2d plot
plt.contour(X, Y, Z, colors='black');
```

```
# attemtping quiver gradient for local minimum
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)

X, Y = np.meshgrid(x, y)
z = X ** 2 + Y ** 2
dx, dy = np.gradient(z)

fig, ax = plt.subplots(figsize=(7,7))
Q = ax.quiver(X,Y,dx,dy, units='width')

qk = ax.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
plt.show()

```

## Function with relative maximum

```
#Function with relative maximum
def max(x, y):
  return -x ** 2 - y ** 2
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

#Plotting in a three-dimensional 3d contour plot first
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = max(X, Y)

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

#Marking local max
ax.scatter(0,0,0, marker='o', color='b')
```

```
#Better view of the local max in the 3d contour plot
ax.view_init(60, 35)
fig
```

```
#Visualizing the function in a 2d plot
plt.contour(X, Y, Z, colors='black');
```

```
# Attempting Quiver of function for local max
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)

X, Y = np.meshgrid(x, y)
z = -X ** 2 - Y ** 2
dx, dy = np.gradient(z)

fig, ax = plt.subplots(figsize=(7,7))
Q = ax.quiver(X,Y,dx,dy, units='width')

qk = ax.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
plt.show()
```

## Function with a point where the gradient is zero

```
def grad(x, y):
  return x ** 2 - y **2
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

#Plotting in a three-dimensional 3d contour plot first
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = grad(X, Y)

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

#Marking saddle point
ax.scatter(0,0,0, marker='o', color='g')
```
